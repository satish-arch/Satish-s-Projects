## Data Cleaning

```python
df = pd.read_csv('Sales.csv')
df = df.dropna(how='all')
df
```

```python
df.info()
```

```python
df = df.loc[  df['Quantity Ordered']!='Quantity Ordered'   ]
df
```

```python
df.info()
```

```python
df.to_csv('Sales.csv',index=False)
df = pd.read_csv('Sales.csv')
df
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Month'] = df['Order Date'].dt.month
df['Month_name'] = df['Order Date'].dt.month_name()
df
```

```python
df['Amount'] = df['Quantity Ordered'] * df['Price Each']
df['City'] = df['Purchase Address'].str.split(', ',expand=True)[1]
df
```

```python
ndf = df.groupby(['Month','Month_name']).agg(
    Total_sale = ('Amount','sum')
)

ndf = ndf.reset_index()
ndf
```

## Plotting Sales by Month

```python
plt.figure(figsize=(12,6))
plt.plot(ndf['Month_name'],ndf['Total_sale'],marker='o',color='blue')
plt.xlabel('Month Name')
plt.ylabel('Sales')
plt.xticks(fontsize=8,rotation=45)
plt.grid()


for x,y in zip(ndf['Month_name'],ndf['Total_sale']):
    plt.text(x,y,y,va='top',ha='left',color='black',fontsize=10)


plt.show()
```

```python
sns.catplot(data=ndf,x='Month_name',y='Total_sale',kind='bar',color='skyblue')
plt.xticks(fontsize=5)
plt.show()
```

```python
ndf = df.groupby('City').agg(
    City_sale = ('Quantity Ordered','sum')
).sort_values('City_sale',ascending=False)
ndf = ndf.reset_index()
ndf
```

# Which city sold the most product?

```python
Max_sale_city = ndf.loc[ndf['City_sale'].idxmax()]
Max_sale_city.reset_index()
```

```python
plt.pie(ndf['City_sale'],labels=ndf['City'],autopct='%.f%%')
plt.show
```

# What time should display advertisements

```python
df['Time'] = df['Order Date'].dt.hour
df
```

```python
ndf = df['Time'].value_counts().reset_index()
ndf.columns = ['Hour','count']
ndf = ndf.sort_values('Hour')

ndf
```

```python

sns.relplot(data=ndf,x='Hour',y='count',kind='line',color='red')
plt.grid()
plt.xticks(np.arange(0,24))
plt.show()
```

## Frequently Bought Together Products

```python
edf = df.loc[df['Order ID'].duplicated(keep=False)]
edf = edf[ ['Order ID','Product'] ]
edf
```

```python
edf['Product'] = edf['Product'] + ','
edf
```

```python
edf = edf.groupby('Order ID').agg(
    Product = ('Product','sum')
)

edf = edf.reset_index()
edf
```

```python
edf['Product'] = edf['Product'].str.strip(',')
edf
```

```python
edf['Product'] = edf['Product'].str.split(',')
edf
```

## Finding Most Common Product Pairs

```python
from itertools import combinations
from collections import Counter

lis = list(edf['Product'])

count = Counter()
for sub_list in lis:
    a = Counter(combinations(sub_list,2))
    count.update(a)
    
count
```
